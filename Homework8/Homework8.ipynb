Получить интуицию, на что влияют параметры распределений. Построить по 5 функций плотностей с разными параметрами для каждого распределения: нормальное, экспоненциальное, Стьюдента на одном графике.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts

%matplotlib inline
```

```python
#нормальное распределение
mu1 = 3.0
sigma1 = 1

mu2 = 2.0
sigma2 = 0.5

mu3 = 1.0
sigma3 = 0.2

mu4 = 3.0
sigma4 = 0.6

mu5 = 2.5
sigma5 = 1.5



# зададим нормально распределенную случайную величину
norm_rv1 = sts.norm(loc=mu1, scale=sigma1)
norm_rv2 = sts.norm(loc=mu2, scale=sigma2)
norm_rv3 = sts.norm(loc=mu3, scale=sigma3)
norm_rv4 = sts.norm(loc=mu4, scale=sigma4)
norm_rv5 = sts.norm(loc=mu5, scale=sigma5)

```

```python
x = np.linspace(0,6,100)
pdf1 = norm_rv1.pdf(x)
pdf2 = norm_rv2.pdf(x)
pdf3 = norm_rv3.pdf(x)
pdf4 = norm_rv4.pdf(x)
pdf5 = norm_rv5.pdf(x)
plt.plot(x, pdf1)
plt.plot(x, pdf2)
plt.plot(x, pdf3)
plt.plot(x, pdf4)
plt.plot(x, pdf5)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
#Экспоненциальное распредление

lam1 = 0.1
lam2 = 0.5
lam3 = 0.1
lam4 = 2.0
lam5 = 3.0

expon_rv1 = sts.expon(lam1)
expon_rv2 = sts.expon(lam2)
expon_rv3 = sts.expon(lam3)
expon_rv4 = sts.expon(lam4)
expon_rv5 = sts.expon(lam5)
```

```python
x = np.linspace(0,5,1000)
сdf1 = expon_rv1.cdf(x)
сdf2 = expon_rv2.pdf(x)
сdf3 = expon_rv3.cdf(x)
сdf4 = expon_rv4.pdf(x)
сdf5 = expon_rv5.cdf(x)
plt.plot(x, сdf1)
plt.plot(x, сdf2)
plt.plot(x, сdf3)
plt.plot(x, сdf4)
plt.plot(x, сdf5)

plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
#распределение стьюдента

n1 = 10
n2 = -10
n3 = 5
n4 = 100
n5 = 50

t_rv1 = sts.t(n1)
t_rv2 = sts.t(n2)
t_rv3 = sts.t(n3)
t_rv4 = sts.t(n4)
t_rv5 = sts.t(n5)

```

```python
x = np.linspace(-5,5,1000)
cdf1 = t_rv1.cdf(x)
cdf2 = t_rv2.cdf(x)
cdf3 = t_rv3.pdf(x)
cdf4 = t_rv4.cdf(x)
cdf5 = t_rv5.pdf(x)

plt.plot(x, cdf1)
plt.plot(x, cdf2)
plt.plot(x, cdf3)
plt.plot(x, cdf4)
plt.plot(x, cdf5)

plt.ylabel('$F(x)$')
plt.xlabel('$x$')
plt.show()
```

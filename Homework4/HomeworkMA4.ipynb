$f(x)=x^5 + 4*sin(2x)+cos(3x+3)  в    точке  $(1)$

$f(x)'=(x^5 + 4*sin(2x)+cos(3x+3))' = (5^x)' + (4*sin(2x))'+ (cos(3x+3))' = 5 * x^4 + 8 * cos(2x) - 3*sin(3x+3)$

В точке 1 производная будет равна $5 * 1^4 + 8 * cos 2 - 3 * sin 6$

Вторая прозводная:

$f(x)'' = (5 * x^4 + 8 * cos(2x) - 3*sin(3x+3))' = 20*x^3 - 16 sin(2x) -9 *cos(3x+3) &

В точке 1 вторая производная будет равна  $20 - 16 sin 2 - 9 * cos 6$

Уровень 1**

Постройте график первой и второй производной функции

(sin(2x+1))^5

на отрезке [-5; 5]


```python
x = np.array([i for i in range(-5, 5)])
y = np.array([math.sin(math.radians(2*x_i+1)**5) for x_i in x])
plt.plot(x, y)
```

```python
#Первая производная
list_der_sin = []
for i in range(len(x)-1):
    list_der_sin.append((y[i+1]-y[i])/(x[i+1]-x[i]))
list_der_sin.append(list_der_sin[-1])
```

```python
fig, axs = plt.subplots(2,figsize=(20,10))
axs[0].set_title('Исходная: sin(2x+1)^5')
axs[0].plot(x, y)
axs[1].set_title('Первая производная функции')
axs[1].plot(x, list_der_sin)
axs[1].plot(x, [0 for i in range(len(list_der_sin))], '-')
```

```python
#вторая производная
list_second_der_sin = []
for i in range(len(x)-1):
    list_second_der_sin.append((list_der_sin[i+1]-list_der_sin[i])/(x[i+1]-x[i]))
list_second_der_sin.append(list_second_der_sin[-1])
```

```python
fig, axs = plt.subplots(3,figsize=(20,10))
axs[0].set_title('Исходная: sin(2x+1)^2')
axs[0].plot(x, y)
axs[1].set_title('Первая производная функции')
axs[1].plot(x, list_der_sin)
axs[1].plot(x, [0 for i in range(len(list_der_sin))], '-')
axs[2].set_title('Вторая производная функции')
axs[2].plot(x, list_second_der_sin)
axs[2].plot(x, [0 for i in range(len(list_der_sin))], '-')
```
